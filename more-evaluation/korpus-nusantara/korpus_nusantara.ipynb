```python
# Batak Toba

bbc_df = pd.read_excel("korpus nusantara.xlsx", sheet_name='Batak')

bbc_df = bbc_df.iloc[:, [1]]
bbc_df.columns = ["text"]

bbc_df["label"] = "bbc"

display(bbc_df.head())

bbc_df.to_csv("bbc.csv", index=False)
```
